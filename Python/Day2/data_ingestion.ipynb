```python
def numeric_year(year_string: str) -> int:
    try:
        year_number_str, year_era_notation_str = year_string.split(" ") # tulemuseks list, kuna järjekord on oluline
    except:
        raise ValueError(f"'{year_string}' is not a valid year number: exactly two components expected.")
    
    try:
        year_numeric = int(year_number_str)
    except:
        raise ValueError(f"'{year_string}' is not a valid year number: year not numeric.")
    
    if year_era_notation_str == "A.D.":
        return int(year_number_str)
    if year_era_notation_str == "B.C.":
        return -int(year_number_str)
    else:
        raise ValueError(f"'{year_string}' is not a valid year number: unknown era.")
    

print(numeric_year("2023 A.D."))
print(numeric_year("1993 B.C."))
```
