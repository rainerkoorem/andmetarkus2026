## if/else

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C")
        if year <= 10000:
            print("Stone Age")
    elif year == 0:
        print("Happy birthday, Jesus")
    else:
        print("A.D")
```

```python
make_sense_of_this_year(2000)
```

```python
def format_year(year: int) -> str:
    if year < 0:
        return f"{abs(year)} B.C"
    else:
        return f"{year} A.D"
print (format_year(2000))
```

```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)": "sepal_length",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_length",
    "petal width (cm)": "petal_width"
})
iris
```

```python
iris.to_csv("iris.csv", index=False)
```

```python
duckdb.sql("""
    SELECT 'Hello, World!'
 """)
```

```python
duckdb.sql("""
    SELECT species, avg(petal_length) AS avg_petal_length
    FROM iris
    GROUP BY species
    ORDER BY avg_petal_length DESC
 """).df()
```

```python
def enne_või_pärast_kristust(aasta: int) -> str:
    if aasta < 0:
        return f"{abs(aasta)} B.C."
    elif aasta == 0:
        return "I believe is Jesus is life this morning."
    else:
        return f"{aasta} A.D."
    
enne_või_pärast_kristust(0)
```

```python
inimesed = ['kati', 'mati', 'silveraadomees', 'hannes']
numbrid = ['Esimese', 'Teise', 'Kolmanda', 'Neljanda']

järjekorranumber = 0

for nimi in inimesed:
    nime_pikkus = len(nimi)
    tekst = f"{numbrid[järjekorranumber]} inimese nimi on {nimi.capitalize()} nime pikkusega {nime_pikkus}"
    if nime_pikkus < 5:
        print(f"{tekst} tähte ning ta kannab lühikest nime.")
    else:
        print(f"{tekst} tähte ning ta kannab pikka nime.")
    järjekorranumber += 1
```

## Ülesanne
Vaja leida ostukorvi koguhind.
- Kui toote hind on > 100, siis soodustus 20%
- Kui toote hind on > 50, siis soodustus 10%

```python
## Raineri variant

hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]
soodustus_suur = 0.8
soodustus_vaike = 0.9
uued_hinnad = []

def get_price(hinnad: list[int]) -> float:
    ostukorvi_hind = 0
    for kaup in hinnad:
        if kaup > 100:
            ostukorvi_hind += (kaup * soodustus_suur)
            uued_hinnad.append(round(kaup * soodustus_suur, 2))
        elif kaup > 50:  
            ostukorvi_hind += (kaup * soodustus_vaike) 
            uued_hinnad.append(round(kaup * soodustus_vaike, 2))
        else:
            ostukorvi_hind += kaup
            uued_hinnad.append(round(kaup, 2)) 
    return ostukorvi_hind, uued_hinnad


## Alternatiiv
alternatiiv_ylesanne = """
def get_discount(price: int) -> float:
    if price > 100:
        return price * soodustus_suur
    elif price > 50:
        return price * soodustus_vaike
    return price

def get_price_alternative(hinnad: list[int]) -> float:
    tulemus = 0
    for hind in hinnad:
        soodushind = get_discount(hind)
        tulemus += soodushind
    return tulemus
"""
get_price(hinnad)
```
